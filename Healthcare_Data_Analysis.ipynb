# Univariate analysis<br>


*   Univariate analysis is the simplest form of data analysis, focusing on one variable at a time to describe its characteristics, find patterns, and summarize data using measures like mean, median, mode, and frequency, often visualized with histograms or bar charts, helping identify outliers and data distribution without looking at relationships between variables.
*   It's thefoundational step in data exploration, essential for understanding individual attributes like height, age, or sales figures before moving to bivariate or multivariate analyses.

# Bivariate Analysis


*   Bivariate analysis is a statistical method to study the relationship between exactly two variables, determining the strength and direction of their association, often using techniques like scatterplots, correlation, or regression to see how one variable changes as the other does (e.g., height and weight).
*    It helps researchers understand patterns, test simple hypotheses, and make predictions by identifying if variables move together (positive), in opposite directions (negative), or have no clear link, differentiating correlation from causation.

## Common Techniques

*  Scatterplots: Visualizes data points to see patterns.
*  Correlation Coefficients (Pearson, Spearman): Quantify the relationship.
*  Cross-Tabulations (Contingency Tables):For categorical variables.
*  T-tests/ANOVA: Compare means between groups (e.g., different genders).


```python
# Gender distribution per medical category

plt.figure(figsize=(8,4))
colors=['yellow','green']
sns.countplot(data=df,x='Medical Condition',hue='Gender',palette=colors,edgecolor='black')
plt.title('Gender Distribution by Medical Category',fontweight='bold')
plt.xlabel('Medical Condition',size=13)
plt.ylabel('Count of Patients by Gender',size=13)
plt.show()
```
